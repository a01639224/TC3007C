```python
# Import the necessary libraries and modules
import pandas as pd
from transformers import pipeline
import nltk
from nltk.tokenize import sent_tokenize
```

## Seleccione un corpus a trabajar.

https://www.kaggle.com/datasets/saurabhshahane/twitter-sentiment-dataset?select=Twitter_Data.csv


```python
# Import the selected Twitter dataset
twitter_df = pd.read_csv('/content/drive/MyDrive/zPLN/Twitter_Data.csv')
twitter_df.head()
```

```python
# Transform column names and true sentiment values
twitter_df.columns = ['tweet', 'true_sentiment']
twitter_df['true_sentiment'] = twitter_df['true_sentiment'].replace({-1:'negative', 0:'neutral', 1:'positive'})
twitter_df.head()
```

## Aplicar un modelo pre entrenado del “sentiment analisis” y analizar reacciones de un post.

https://huggingface.co/cardiffnlp/twitter-roberta-base-sentiment

```python
# Choose a sentiment analysis model from the ones available
sentiment_analysis_model = pipeline(model='cardiffnlp/twitter-roberta-base-sentiment')
```

```python
# Predict the sentiment of five Tweets from the database using the selected model
sample = twitter_df.sample(n=5)
predictions = sentiment_analysis_model(sample['tweet'].values.tolist())
sample['predicted_sentiment'] = [i['label'] for i in predictions]
sample['score'] = [i['score'] for i in predictions]
sample['predicted_sentiment'] = sample['predicted_sentiment'].replace({'LABEL_0':'negative', 'LABEL_1':'neutral', 'LABEL_2':'positive'})
sample
```

## Separe las oraciones después de cada punto. Contabilice el número total de oraciones.

```python
# Download Punkt tokenizer models for splitting Tweets into sentences
nltk.download('punkt')
```

```python
# Count and compare the number of sentences in the database
count = 0

for i in twitter_df['tweet'].values.tolist():
  count = count + len(sent_tokenize(str(i)))

print('No. of Tweets:', twitter_df.shape[0])
print('No. of sentences:', count)
```
